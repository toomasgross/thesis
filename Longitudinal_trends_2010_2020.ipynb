```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.ticker as mtick  
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
# read the files

df_2020_original = pd.read_csv("C:/Users/Toomas/Desktop/ITMI_Data/Cleaned_datasets_25_64/Yearly/2020_original_cleaned.csv")
df_2020 = pd.read_csv("C:/Users/Toomas/Desktop/ITMI_Data/Cleaned_datasets_25_64/Yearly/2020_cleaned.csv")
df_2004_2020 = pd.read_csv("C:/Users/Toomas/Desktop/ITMI_Data/Cleaned_datasets_25_64/Joined/2004_2020_cleaned.csv")
```

```python
# see bmi means by year of survey

df_2004_2020.groupby('survey_year').bmi.mean()
```

```python
# plot of average BMI per survey year

list_bmi_means = []
year = 2004
print('Population percentage exercising more than once a week by survey year: ')

for i in range(1,10):
    
    df_temp = df_2004_2020[df_2004_2020.survey_year == year]
    list_bmi_means.append(df_temp.bmi.mean())
    year = year +2

years = ['2004','2006','2008','2010','2012','2014','2016','2018','2020']

plt.bar(years,list_bmi_means)
plt.savefig("C:/Users/Toomas/Desktop/ITMI_Data/Kood/MSc/Plots/EDA1.pdf")
    
```

```python
# value counts for bmi_two_groups_split25 by survey year

df_2004_2020.groupby('survey_year').bmi_two_groups_split25.value_counts()

```

```python
# print out the percentage of overweight + obese respondents from all yearly respondents by survey year

df_2004_2020_temp = df_2004_2020[df_2004_2020.bmi_two_groups_split25 == 2]

list_25more = []

year = 2004
print('Overweight/obese population percentage by survey year: ')

for i in range(1,10):
    yearly_sample_size = (df_2004_2020.survey_year == year).sum()
    yearly_overweight_individual_count = (df_2004_2020_temp.survey_year == year).sum()
    yearly_percentage_overweight= round(yearly_overweight_individual_count*100/yearly_sample_size,1)
    print(year, ': ', yearly_percentage_overweight)
    list_25more.append(yearly_percentage_overweight)
    year = year +2
```

```python
# print out the percentage of obese respondents from all yearly respondents by survey year

df_2004_2020_temp = df_2004_2020[df_2004_2020.bmi_two_groups_split30 == 2]

list_30more = []
year = 2004
print('Obese population percentage by survey year: ')

for i in range(1,10):
    yearly_sample_size = (df_2004_2020.survey_year == year).sum()
    yearly_obese_individual_count = (df_2004_2020_temp.survey_year == year).sum()
    yearly_percentage_obese= round(yearly_obese_individual_count*100/yearly_sample_size,1)
    print(year, ': ', yearly_percentage_obese)
    list_30more.append(yearly_percentage_obese)
    year = year +2
```

```python
# print out the percentage of those respondents who exercise more than once a week
# (from all yearly respondents by survey year)

df_2004_2020_temp = df_2004_2020[df_2004_2020.exercising_frequency > 4]

list_exercise = []
year = 2004
print('Population percentage exercising more than once a week by survey year: ')

for i in range(1,10):
    yearly_sample_size = (df_2004_2020.survey_year == year).sum()
    yearly_exercisers_individual_count = (df_2004_2020_temp.survey_year == year).sum()
    yearly_percentage_exercisers= round(yearly_exercisers_individual_count*100/yearly_sample_size,1)
    print(year, ': ', yearly_percentage_exercisers)
    list_exercise.append(yearly_percentage_exercisers)
    year = year +2
```

```python
# print out the percentage of those who consume alcohol weekly

df_2004_2020_temp = df_2004_2020[df_2004_2020.alcohol_standard_units_consumption_frequency > 3]

list_alcohol = []

year = 2004
print('Weekly alcohol consumption percentage by survey year: ')

for i in range(1,10):
    yearly_sample_size = (df_2004_2020.survey_year == year).sum()
    yearly_alcohol_count = (df_2004_2020_temp.survey_year == year).sum()
    yearly_percentage_alcohol= round(yearly_alcohol_count*100/yearly_sample_size,1)
    print(year, ': ', yearly_percentage_alcohol)
    list_alcohol.append(yearly_percentage_alcohol)
    year = year +2
```

```python
# print out the percentage of those who smoke

df_2004_2020_temp = df_2004_2020[df_2004_2020.smoking_history > 2]

list_smoking = []

year = 2004
print('Smoking by survey year: ')

for i in range(1,10):
    yearly_sample_size = (df_2004_2020.survey_year == year).sum()
    yearly_smoking_count = (df_2004_2020_temp.survey_year == year).sum()
    yearly_percentage_smoking= round(yearly_smoking_count*100/yearly_sample_size,1)
    print(year, ': ', yearly_percentage_smoking)
    list_smoking.append(yearly_percentage_smoking)
    year = year +2
```

```python
# make a new dataframe with survey years 2004-2020 and the three yearly percentages calculated above 
list_years = [2004,2006,2008,2010,2012,2014,2016,2018,2020]
df_percentages=pd.DataFrame(list(zip(list_years,list_25more,list_30more,list_exercise,list_alcohol,list_smoking)), columns=['Survey year','% BMI ≥ 25.0','% BMI ≥ 30.0','% weekly exercising times > 1', '% alcohol consumption at least weekly','% smokers'])
```

```python
# plot this data for the years 2004-2020

import matplotlib.ticker as mtick

lines = df_percentages[['% BMI ≥ 25.0','% BMI ≥ 30.0','% weekly exercising times > 1','% alcohol consumption at least weekly','% smokers']]


ax = lines.plot(kind='line',figsize=(10,10))
ax.grid(axis ='y')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.legend(loc='center right')
ax.set_ylabel('Percentage',size=14)
ax.set_xlabel('Year', size=14)
ax.set_title('Yearly percentages of BMI groups and exercising, 2004-20')


positions=(0,1,2,3,4,5,6,7,8)
new_xticks=('2004','2006','2008','2010','2012','2014','2016','2018','2020')
plt.xticks(positions,new_xticks)

plt.savefig("C:/Users/Toomas/Desktop/ITMI_Data/Kood/MSc/Plots/EDA2_25_64.pdf")
```
